```python
drought['Date'] = pd.to_datetime(drought['Date'])
```

```python
drought.dtypes
```

```python
# Weighting drought categories to reflect relative intensity of higher drought categories in drought factor (drought intensity) 
```

```python
drought['Drought_magnitude'] = ((drought['D0'] * 1) + 
                                (drought['D1'] * 2) + 
                                (drought['D2'] * 3) + 
                                (drought['D3'] * 4) + 
                                (drought['D4'] * 5))
```

```python
drought['Drought_intensity'] = (drought['Drought_magnitude'] / 5)
```

```python
drought.head()
```

```python
drought.max()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
```

```python
# Checking to see of drought factor(intensity) comports with the source data. It does. 
```

```python
sb.lineplot(x="Date", y="Drought_intensity", data=drought)

# set label
plt.ylabel("Drought")

plt.show()
```
